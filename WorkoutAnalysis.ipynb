```python
def longest_streak(steps_data, date_data):
    current_streak = 0
    longest_streak = 0
    start_date_longest = None
    end_date_longest = None
    start_date_current = None

    for i, steps in enumerate(steps_data):
      if steps >= 10000:
        current_streak += 1
        if current_streak == 1:
          start_date_current = date_data[i]
      else:
        if current_streak > longest_streak:
          longest_streak = current_streak
          start_date_longest = start_date_current
          end_date_longest = date_data[i - 1]

        current_streak = 0

    if current_streak > longest_streak:
          longest_streak = current_streak
          start_date_longest = start_date_current
          end_date_longest = date_data[-1]

    return longest_streak, start_date_longest, end_date_longest

steps_data = data['steps'].tolist()
date_data = data['date'].tolist()

longest_streak_count, start_date, end_date = longest_streak(steps_data, date_data)

print("Longest Streak = ", longest_streak(steps_data, date_data))
print("Start Date of Longest Streak:", start_date)
print("End Date of Longest Streak:", end_date)
```

```python
def longest_streak(steps_data, date_data):
    streak_info = []  # List to store streak information

    current_streak = 0
    start_date_current = None

    for i, steps in enumerate(steps_data):
        if steps >= 10000:
            current_streak += 1
            if current_streak == 1:
                start_date_current = date_data[i]
        else:
            if current_streak > 0:  # If a streak has ended
                streak_info.append({
                    'Streak': current_streak,
                    'Start Date': start_date_current,
                    'End Date': date_data[i - 1]
                })
            current_streak = 0

    # Check if the current streak at the end of the data is longer than the longest streak, just to be sure
    if current_streak > 0:
        streak_info.append({
            'Streak': current_streak,
            'Start Date': start_date_current,
            'End Date': date_data[-1]
        })

    # Convert streak_info list to DataFrame and sort it by 'Streak'
    streak_df = pd.DataFrame(streak_info).sort_values(by='Streak', ascending=False)

    return streak_df

# Example usage
steps_data = data['steps'].tolist()
date_data = data['date'].tolist()

streak_table = longest_streak(steps_data, date_data)
print(streak_table)
```
